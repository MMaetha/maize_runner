```python
%matplotlib inline
from collections import Counter;
```

```python
f = open('TE_search07_chr1.csv','r')
data = f.readlines()
f.close()
data2 = [int(n) for n in data] #mapping by list comprehenson
data3 = [n for n in data2 if n>=0] #filtering data by list comprehenson
num = Counter(data3)
print(num)
```

```python
data4 = []  
tmp=0;

for n in range(0,10000):
    tmp = max(num[n],tmp)
    if(n%50==0):
        data4.append(tmp)
        tmp=0

print(data4)
```

```python
import matplotlib.pyplot as plt
```

```python
N = len(data4)
x = range(N)
width = 1/1.5
plt.bar(x, data4, width, color="blue")
```

```python
data5 = []  
sum = 0;

for n in range(0,10000):
    sum+=num[n];
    if((n+1)%50==0):
        sum/=50
        data5.append(sum)
        sum = 0

print(data5)
```

```python
N = len(data5)
x = range(N)
width = 1/1.5
plt.bar(x, data5, width, color="blue")
```
